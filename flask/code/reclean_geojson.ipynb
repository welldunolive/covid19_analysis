```python
import pandas as pd
import numpy as np
import requests
import folium
import json
```

```python
ca = pd.read_csv('ca_counties.csv')
ca.sort_values(by = ['county'], inplace = True)
ca
```

```python
with open('ca-counties.geojson') as f:
    json_counties = json.load(f)

tooltip_text = []
for idx in range(len(ca)):
    tooltip_text.append(str(round(ca['pop_density'][idx])))

for idx in range(len(tooltip_text)):
        json_counties['features'][idx]['properties']['Pop'] = tooltip_text[idx]
        
        
tooltip_text2 = []
for idx in range(len(ca)):
    tooltip_text2.append('$' + str(ca['inc_per_capita_inc_dol'][idx]))

for idx in range(len(tooltip_text2)):
        json_counties['features'][idx]['properties']['Income'] = tooltip_text2[idx]
        
        
tooltip_text3 = []
for idx in range(len(ca)):
    tooltip_text3.append(str(round(ca['percent_health_ins_noninst_pop_cov_yes'][idx] * 100)) +'%')

for idx in range(len(tooltip_text3)):
        json_counties['features'][idx]['properties']['Insurance'] = tooltip_text3[idx]
        
        
tooltip_text4 = []
for idx in range(len(ca)):
    tooltip_text4.append(str(round(ca['percent_sex_age_pop_male'][idx] * 100)) + '%')

for idx in range(len(tooltip_text4)):
        json_counties['features'][idx]['properties']['Sex'] = tooltip_text4[idx]
        
        
tooltip_text5 = []
for idx in range(len(ca)):
    tooltip_text5.append(str(round(ca['percent_sex_age_pop_45_to_74'][idx] * 100)) + '%')

for idx in range(len(tooltip_text5)):
        json_counties['features'][idx]['properties']['Age'] = tooltip_text5[idx]
        
        
tooltip_text6 = []
for idx in range(len(ca)):
    tooltip_text6.append(str(int(round(ca['tests_per_100_people'][idx]))))

for idx in range(len(tooltip_text6)):
        json_counties['features'][idx]['properties']['Test'] = tooltip_text6[idx]
        
        
tooltip_text7 = []
for idx in range(len(ca)):
    tooltip_text7.append(str(round(ca['obes_percent'][idx] * 100)) + '%')

for idx in range(len(tooltip_text7)):
        json_counties['features'][idx]['properties']['Obesity'] = tooltip_text7[idx]
```

```python
with open('ca-counties2.geojson', 'w', encoding='utf-8') as f:
    json.dump(json_counties, f, ensure_ascii=False)
```
